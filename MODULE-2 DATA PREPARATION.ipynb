```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
from glob import glob
```

```python
femalepath=glob('./DATA/female/*.jpg')
malepath=glob('./DATA/male/*.jpg')
```

```python
len(femalepath),len(malepath)
```

```python
path=femalepath[0]
img=cv.imread(path)
img.shape
```

```python
def extract_images(path,gender,i) :
    img=cv.imread(path)
    gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
    haar=cv.CascadeClassifier('./DATA/haarcascade_frontalface_default.xml')
    faces=haar.detectMultiScale(gray,1.5,5)
    for x,y,w,h in faces:
        roi=img[y:y+h,x:x+w]
        if(gender=='male'):
            cv.imwrite('./DATA/crop/male_crop/{}_{}.png'.format(gender,i),roi)
        else:
            cv.imwrite('./DATA/crop/female_crop/{}_{}.png'.format(gender,i),roi)
```

```python
for i,path in enumerate(malepath):
    try:
        extract_images(path,'male',i)
        print('INFO: {}/{} processed successfully'.format(i,len(femalepath)))
    except:
        print('INFO: {}/{} cannot be processed'.format(i,len(femalepath)))
```

```python
for i,path in enumerate(femalepath):
    try:
        extract_images(path,'female',i)
        print('INFO: {}/{} processed successfully'.format(i,len(femalepath)))
    except:
        print('INFO: {}/{} cannot be processed'.format(i,len(femalepath)))
```
