```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
from glob import glob
```

```python
#load numpy zip
data=np.load('./DATA/data_100_100_norm.npz')
```

```python
X=data['arr_0'] #independent features
y=data['arr_1'] # target value
```

```python
X.shape,y.shape
```

# Eigen Image

```python
X1=X-X.mean(axis=0)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=None,whiten=True,svd_solver='auto')
```

```python
x_pca=pca.fit_transform(X1)
```

```python
x_pca.shape
```

```python
eigen_ratio=pca.explained_variance_ratio_
eigen_ratio_cum=np.cumsum(eigen_ratio)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(eigen_ratio[:200],'r>--')
plt.xlabel("No of components")
plt.ylabel("Explained Variance ratio")
plt.subplot(1,2,2)
plt.xlabel("No of components")
plt.ylabel("Cummulative sum of the explained Variance ratio")
plt.plot(eigen_ratio_cum[:200],'r>--')
```

conclusion: using elbow method , consider number of component in between 25-30
> sinve if i consider between 25-30 the explained variance is around 80% so, in order
to get min 85% variance i am considering 50 components

```python
pca_50=PCA(n_components=50,whiten=True,svd_solver='auto')
x_pca_50=pca_50.fit_transform(X1)
```

```python
x_pca_50.shape
```

```python
#saving pca
import pickle
```

```python
pickle.dump(pca_50,open('./model/pca_50.pickle','wb'))
```

```python
# consider 50 component and inverse transform 
x_pca_inv=pca_50.inverse_transform(x_pca_50)
```

```python
x_pca_inv.shape
```

```python
# consider one image (one row)
eig_img=x_pca_inv[0,:]
eig_img=eig_img.reshape((100,100))
plt.imshow(eig_img,cmap='gray')
```

```python
def label(y):
    if y==0:
        return 'male'
    return 'female'
np.random.randint(1001)
pics=np.random.randint(0,438,40)
plt.figure(figsize=(15,8))
for i,pic in enumerate(pics):
    plt.subplot(4,10,i+1)
    img=X[pic:pic+1].reshape(100,100)
    plt.imshow(img,cmap='gray')
    plt.title('{}'.format(label(y[pic])))
    plt.xticks([])
    plt.yticks([])
plt.show()
print("="*20+'Eigen Images'+"="*20)
plt.figure(figsize=(15,8))
for i,pic in enumerate(pics):
    plt.subplot(4,10,i+1)
    img=x_pca_inv[pic:pic+1].reshape(100,100)
    plt.imshow(img,cmap='gray')
    plt.title('{}'.format(label(y[pic])))
    plt.xticks([])
    plt.yticks([])
plt.show()
```
