```python
cap=cv.VideoCapture('C:\\Users\\shekh\\Downloads\\CN\\Application Layer Track - Practice - GeeksforGeeks.ts')
```

```python
def detectface(img):
    gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
    haar=cv.CascadeClassifier('./DATA/haarcascade_frontalface_default.xml')
    faces=haar.detectMultiScale(gray)
    for x,y,w,h in faces:
        cv.rectangle(img,(x,y),(x+w,y+h),(0,255,255),3)
    return img
```

```python
while True:
    ret,frame=cap.read()
    if ret==False:
        break
#     gray=cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
#     cv.imshow('gray',gray)
    frame=detectface(frame)
    cv.imshow('object Detected',frame)
    if cv.waitKey(40)==27:
        break
cv.destroyAllWindows()
cap.release()
```
